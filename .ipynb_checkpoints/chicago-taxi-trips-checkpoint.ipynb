```python
import warnings
warnings.filterwarnings('ignore')

from google.cloud import bigquery
import pandas as pd

client = bigquery.Client(project='project-b7e71fb0-039e-408e-a0e')

query = """
SELECT 
    company,
    COUNT(*) AS trips,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (), 2) AS market_share_pct
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE EXTRACT(YEAR FROM trip_start_timestamp) = 2023
GROUP BY company
ORDER BY trips DESC
LIMIT 10
"""

df = client.query(query).to_dataframe()
print(df)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# График 1 — количество поездок
axes[0].barh(df['company'], df['trips'], color='steelblue')
axes[0].set_title('Топ-10 компаний по поездкам')
axes[0].set_xlabel('Количество поездок')
axes[0].invert_yaxis()

# График 2 — доля рынка
axes[1].pie(df['market_share_pct'], labels=df['company'], autopct='%1.1f%%')
axes[1].set_title('Доля рынка')

plt.tight_layout()
plt.savefig('chicago_taxi_analysis.png')
plt.show()
```
